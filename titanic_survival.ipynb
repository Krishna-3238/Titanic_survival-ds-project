```python
import pandas as pd
```

```python
train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
train_df.info()
```

```python
train_df.isnull().sum()
```

```python
train_df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
train_df.head()
```

```python
train_df.Sex = train_df.Sex.map({'male':1,'female':2})
```

```python
train_df.Age = train_df.Age.fillna(train_df.Age.mean())
```

```python
train_df.head(10)
```

```python
x_train = train_df.drop('Survived',axis='columns')
y_train = train_df.Survived
```

```python
x_train.isnull().sum()
```

```python
test_df = pd.read_csv('test.csv')
test_df.head()
```

```python
test_df.info()
```

```python
test_df.isnull().sum()
```

```python
test_df.drop(['Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
test_df.head(10)
```

```python
test_df.Sex = test_df.Sex.map({'male':1,'female':2})
```

```python
test_df.Age = test_df.Age.fillna(test_df.Age.mean())
```

```python
test_df.Fare = test_df.Fare.fillna(test_df.Fare.mean())
```

```python
test_df.isnull().sum()
```

```python
x_test = test_df.iloc[:,1:]
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression() 
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
y_pred
```

```python
data1 = pd.DataFrame(test_df.PassengerId)
data1
```

```python
data = pd.DataFrame(y_pred,columns=['Survived'])
data
```

```python
result = pd.concat([data1,data],axis=1)
result
```
